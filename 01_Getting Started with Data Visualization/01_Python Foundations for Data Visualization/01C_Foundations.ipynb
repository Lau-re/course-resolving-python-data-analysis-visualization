## Create objects with data

### Groceries

```python
groceries = ['Pan', 'Tomate', 'Vino']
```

### Price

```python
prices = [1, 1.5, 3]
```

## Import matplotlib library

> **Mat**hematical **Plot**ting **Lib**rary

Matplotlib is a Python library that contains functions to create Data Visualizations

```python
import matplotlib.pyplot as plt
```

### Scatter plot (points)

```python
plt.scatter(x=groceries, y=prices)
```

### Bar plot

```python
plt.bar(x=groceries, height=prices)
```

### Many others

https://matplotlib.org/stable/plot_types/index

## Customize the plot

### Change the size

```python
plt.scatter(x=groceries, y=prices, s=75)
```

### Change the color

```python
plt.scatter(x=groceries, y=prices, s=75, c='red')
```

### Change the marker

https://matplotlib.org/stable/api/markers_api.html

```python
plt.scatter(x=groceries, y=prices, s=75, c='red', marker='+')
```

### Change the border

```python
plt.scatter(x=groceries, y=prices, s=300, c='red', marker='o', edgecolors='blue')
```

```python
plt.scatter(x=groceries, y=prices, s=200, c='orange', marker='X', edgecolors='blue')
```
